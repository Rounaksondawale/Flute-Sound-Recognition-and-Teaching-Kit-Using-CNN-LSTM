```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
array1 = np.load(r"C:\Resume_projects\flute_teaching_kit\data\processed_data\sa\sa_81.npy")

fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(array1,
                               x_axis='time',
                               y_axis='log',
                               ax=ax)

ax.set_title('mel spectogram of sa sample', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
```
